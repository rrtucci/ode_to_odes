\\
\dot{y}  = y (1-x^2 -y^2)
\end{array}
\right.
$

```python
def f(t, y, c):
    d2 = 1-y[0]**2 - y[1]**2
    ydot = [y[0]*d2, y[1]*d2]
    return ydot

def dx_dt(x, y):
    return f(0, (x, y), c)[0]
    
def dy_dt(x, y):
    return f(0, (x, y), c)[1]

c=0
```
