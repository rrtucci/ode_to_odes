$
\left\{\begin{array}{l}
\dot{\theta}= v
\\
\dot{v} = - sin(\theta)
\end{array}
\right.$

```python
from ode_helpers import *
from phaseportrait import PhasePortrait2D
```

```python
def ydot(theta, v):
    return v, -np.sin(theta)

pp = PhasePortrait2D(
    ydot, [-9, 9], MeshDim=21, Title='Nonlinear pendulum', xlabel=r"$\Theta$", ylabel=r"$\dot{\Theta}$")
pp.plot()
```

```python
def f(t, y, c):
    return [y[1], -np.sin(y[0])]

def dx_dt(x, y):
    return f(0, (x, y), c)[0]
    
def dy_dt(x, y):
    return f(0, (x, y), c)[1]

c=0
```

```python
plot_phase_plane(dx_dt,
                 dy_dt,
                 x_range=(-9, 9),
                 y_range=(-4, 4),
                 density=20,
                 scale=10
                 # %trajectories=[(2, 0), (-2, 1), (0, -2)]
                 )
```
