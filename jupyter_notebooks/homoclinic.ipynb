# Homoclinic Orbit

$
\left\{
\begin{array}{l}
\dot{x}=y
\\
\dot{y} = x - x^3
\end{array}
\right.
$

```python
def f(t, y, c):
    ydot = [y[1],
         y[0] - y[0]**3]
    return ydot

def dx_dt(x, y):
    return f(0, (x, y), c)[0]
    
def dy_dt(x, y):
    return f(0, (x, y), c)[1]

c= 0
```
